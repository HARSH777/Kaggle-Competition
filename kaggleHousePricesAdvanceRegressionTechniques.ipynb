```python
final_df = pd.concat([trainDataset, test_df], axis = 0)
```

```python
final_df['SalePrice']
```

```python
def category_onehot_multcols(categoricalFeatures):
    df_final=final_df.copy()
    i=0
    for fields in categoricalFeatures:
        
        df1=pd.get_dummies(df_final[fields],drop_first=True)
        
        df_final.drop([fields],axis=1,inplace=True)
        if i==0:
            mergedDummies=df1.copy()
        else:
            
            mergedDummies = pd.concat([mergedDummies,df1],axis=1)
        i=i+1
       
        
    df_final=pd.concat([df_final,mergedDummies],axis=1)
        
    return df_final
```

```python
final_df = category_onehot_multcols(categoricalFeature)
```

```python
final_df.shape
```

```python
duplicated_columns = pd.Series(final_df.columns.duplicated(), index=final_df.columns)
```

```python
duplicateCol = pd.DataFrame(duplicated_columns,columns=['IsDuplicate'])
print(type(duplicateCol['IsDuplicate']))
duplicateCol[duplicateCol['IsDuplicate'] ==True].head()
```

```python
df_Train=final_df.iloc[:1422,:]
df_Test=final_df.iloc[1422:,:]
```

```python
import xgboost
```

```python
regressor = xgboost.XGBRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = dict(
            booster=['gbtree','gblinear'],
            base_score=[0.25,0.5,0.75,1],
            n_estimators = [100, 500, 900, 1100, 1500],
            max_depth = [2, 3, 5, 10, 15],
            learning_rate=[0.05,0.1,0.15,0.20],
            min_child_weight=[1,2,3,4])
```

```python
hyperParameterModel = RandomizedSearchCV(regressor, param_distributions=params, cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
hyperParameterModel.fit(df_Train.drop(['SalePrice'], axis=1), df_Train['SalePrice'])
```

```python
hyperParameterModel.best_estimator_
```

```python
finalModel = xgboost.XGBRegressor(base_score=0.25, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0,
             importance_type='gain', learning_rate=0.1, max_delta_step=0,
             max_depth=2, min_child_weight=1, missing=None, n_estimators=900,
             n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=1, verbosity=1)
```

```python
finalModel.fit(df_Train.drop(['SalePrice'], axis =1), df_Train['SalePrice'])
```

```python
y_pred = finalModel.predict(df_Train.drop(['SalePrice'], axis =1))
```
